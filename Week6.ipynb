**5. Make an image histogram such as the output equalization histogram and also the image display before and after the following equalization histogram process based on the flowchart below : (Use image lena_lc .jpg)**

```python
import cv2 as cv
import matplotlib.pyplot as plt

# Load an image
image_path = '/content/drive/MyDrive/Polinema/Semester 5/PVCK/Meeting 6/lena_lc.jpg'  # Replace with the path to your image
image = cv.imread(image_path)

# Check if the image was loaded successfully
if image is None:
    print("Error: Unable to load the image.")
else:
    # Split the image into RGB channels
    b, g, r = cv.split(image)

    # Apply histogram equalization to each channel
    equalized_r = cv.equalizeHist(r)
    equalized_g = cv.equalizeHist(g)
    equalized_b = cv.equalizeHist(b)

    # Merge the equalized channels to create the equalized color image
    equalized_image = cv.merge([equalized_b, equalized_g, equalized_r])

    # Plot the original and equalized images
    plt.figure(figsize=(12, 6))

    # Original Image
    plt.subplot(1, 2, 1)
    plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
    plt.title('Original Image')
    plt.axis('off')

    # Equalized Image
    plt.subplot(1, 2, 2)
    plt.imshow(cv.cvtColor(equalized_image, cv.COLOR_BGR2RGB))
    plt.title('Equalized Image')
    plt.axis('off')

    plt.tight_layout()
    plt.show()

    # Plot histograms for each RGB channel
    plt.figure(figsize=(12, 4))

    # Original Image Histograms
    plt.subplot(131)
    plt.hist(r.ravel(), bins=256, range=(0, 256), color='red', alpha=0.6)
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')
    plt.title('Red Channel Histogram (Original)')

    plt.subplot(132)
    plt.hist(g.ravel(), bins=256, range=(0, 256), color='green', alpha=0.6)
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')
    plt.title('Green Channel Histogram (Original)')

    plt.subplot(133)
    plt.hist(b.ravel(), bins=256, range=(0, 256), color='blue', alpha=0.6)
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')
    plt.title('Blue Channel Histogram (Original)')

    plt.tight_layout()
    plt.show()

    # Plot histograms for each RGB channel after equalization
    plt.figure(figsize=(12, 4))

    plt.subplot(131)
    plt.hist(equalized_r.ravel(), bins=256, range=(0, 256), color='red', alpha=0.6)
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')
    plt.title('Red Channel Histogram (Equalized)')

    plt.subplot(132)
    plt.hist(equalized_g.ravel(), bins=256, range=(0, 256), color='green', alpha=0.6)
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')
    plt.title('Green Channel Histogram (Equalized)')

    plt.subplot(133)
    plt.hist(equalized_b.ravel(), bins=256, range=(0, 256), color='blue', alpha=0.6)
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')
    plt.title('Blue Channel Histogram (Equalized)')

    plt.tight_layout()
    plt.show()
```

**7. Do the mapping process of 16 million RGB colors into just 8 colors, namely black, green, yellow, blue, cyan, magenta, white as shown in the following output based on the flowchart below! (Use the image testlena .jpg)**

```python
import cv2 as cv
import numpy as np
from sklearn.cluster import KMeans

# Load the original image
image_path = '/content/drive/MyDrive/Polinema/Semester 5/PVCK/Meeting 6/lena_lc.jpg'  # Replace with the path to your image
image = cv.imread(image_path)

# Check if the image was loaded successfully
if image is None:
    print("Error: Unable to load the image.")
else:
    # Reshape the image to a 2D array of pixels
    pixels = image.reshape((-1, 3))

    # Define the number of clusters (in this case, 8 colors)
    num_clusters = 8

    # Perform k-means clustering
    kmeans = KMeans(n_clusters=num_clusters)
    kmeans.fit(pixels)

    # Define the 8 specific colors
    color_map = np.array([
        [0, 0, 0],      # Black
        [0, 255, 0],    # Green
        [255, 255, 0],  # Yellow
        [0, 0, 255],    # Blue
        [0, 255, 255],  # Cyan
        [255, 0, 255],  # Magenta
        [255, 255, 255],# White
        [128, 128, 128] # Gray
    ], dtype=int)

    # Replace each pixel with its corresponding cluster center color
    new_pixels = color_map[kmeans.labels_]

    # Reshape the new pixels back to the original image shape
    quantized_image = new_pixels.reshape(image.shape)

    # Display the original and quantized images
    cv.imshow('Original Image', image)
    cv.imshow('Quantized Image', quantized_image)
    cv.waitKey(0)
    cv.destroyAllWindows()
```
